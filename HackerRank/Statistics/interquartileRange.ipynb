```python
"""
sample input
6
6 12 8 10 20 16
5 4 3 2 1 5
"""

n = int(input())
xarr = list(map(int, input().strip().split(" ")))
farr = list(map(int, input().strip().split(" ")))

sarr = []

for i in range(0, n):
    for f in range(0, farr[i]):
        sarr.append(xarr[i])

            
print(sarr)
# check
if len(sarr)!= sum(farr):
    print("err")
else:
    print("passed check")

    
sorted_arr = sorted(sarr)
arrsize = len(sorted_arr)
print("sorted array: ", sorted_arr)

def findMedian(sortedlist, listsize):
    if listsize%2 == 0:
        middle_lower = int((listsize / 2)) - 1
        middle_higher = int(listsize / 2)
        median = (sortedlist[middle_lower] + sortedlist[middle_higher])/2
    
    else:
        middle = int((listsize / 2))
        median = sortedlist[middle]
        
    return float(median)

q2 = findMedian(sorted_arr, arrsize)


lowerhalf = []
upperhalf = []
# generate lower and upper half
if arrsize%2 == 0:
    mid_index = int((arrsize / 2))
    print("mid index for even case", mid_index)
    lowerhalf = sorted_arr[0:mid_index]
    upperhalf = sorted_arr[mid_index : arrsize]
else:
    mid_index = int(arrsize/2)
    lowerhalf = sorted_arr[0:mid_index]
    upperhalf = sorted_arr[(mid_index + 1) : arrsize]
    
print("lowerhalf: ",lowerhalf)
print("upperhalf:", upperhalf)

q1 = round(findMedian(lowerhalf, len(lowerhalf)),1)
q3 = round(findMedian(upperhalf, len(upperhalf)),1)

# output

# print(q1)
# print(q2)
# print(q3)    
interQrange = round(q3 - q1, 1)
print(interQrange)

```
